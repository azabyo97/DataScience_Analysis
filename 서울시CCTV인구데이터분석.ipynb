```python
# CCTV csv파일 가져오기
csvPath='/content/drive/MyDrive/DataScience/data/01. CCTV_in_Seoul.csv'
xlsxPath='/content/drive/MyDrive/DataScience/data/01. population_in_Seoul.xls'
```

```python
import pandas as pd
import numpy as np
```

```python
cctv_seoul = pd.read_csv(csvPath, encoding='utf-8')

# cctv_seoul의 데이터타입은 데이터프레임
```

```python
# 컬럼명 '기관명' -> '구별' 변경
cctv_seoul.rename(columns={'기관명':'구별'},inplace=True)
cctv_seoul.head()
```

```python
cctv_seoul['최근증가율']=(cctv_seoul['2016년']+cctv_seoul['2015년']+cctv_seoul['2014년'])/cctv_seoul['2013년도 이전']*100
```

```python
# 데이터에 질문을 던지자!
# 가장 CCTV대수가 많은 구는 어디? 적은 구는 어디?
# 소계를 기준으로 가장 많은 구 5개, 적은 구 5개 -> 정렬
cctv_seoul.sort_values(by='소계', ascending=False).head(10) # 소계를 기준으로 가장 많은 구 5개
```

```python
cctv_seoul.sort_values(by='소계', ascending=True).head() # 소계를 기준으로 가장 적은 구 5개
```

```python
# 질문 : 최근 증가율이 가장 높은 구 5개
cctv_seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
# 질문 : 최근 증가율이 가장 낮은 구 5개
cctv_seoul.sort_values(by='최근증가율', ascending=True).head()
```

```python
# cctv_seoul에서 '2013년도 이전', '2014년', '2015년', '2016년' 컬럼을 삭제
cctv_seoul.drop(['2013년도 이전', '2014년', '2015년', '2016년'], axis=1, inplace=True)
cctv_seoul.head()
```

```python
cctv_seoul.tail()
```

```python
# usecols를 알기 위해서는 먼저 excel 파일의 내용을 눈으로 확인한 후에 선택한다.
pop_seoul = pd.read_excel(xlsxPath, header=2, usecols='B, D, G, J, N')
pop_seoul.head()
```

```python
pop_seoul.tail()
```

```python
pop_seoul.columns
```

```python
# 컬럼명 변경하기
pop_seoul.columns = ['구별', '인구', '한국인', '외국인', '고령자']
pop_seoul.head()
```

```python
# 불필요한 행[0, 25]을 삭제하기
pop_seoul.drop([0, 26], inplace=True)
pop_seoul.head()
```

```python
pop_seoul
```

```python
# index 번호를 0~24로 변경
pop_seoul.reset_index(drop=True, inplace=True)
pop_seoul
```

```python
# cctv_seoul, pop_seoul 데이터프레임 합치기
data_result=pd.merge(cctv_seoul, pop_seoul, on='구별')
data_result.head()
```

```python
# 고령자의 비율
data_result['고령자비율']=data_result['고령자']/data_result['인구']*100
```

```python
# 외국인의 비율
data_result['외국인비율']=data_result['외국인']/data_result['인구']*100
```

```python
data_result.columns
```

```python
# 컬럼명 다시 변경
data_result.columns=['구별', 'CCTV대수', '최근증가율', '인구', '한국인', '외국인', '고령자', '고령자비율', '외국인비율']
data_result.head()
```

```python
# matplotlib으로 그래프를 그리기 위해서는 index를  '구별'로 변경
data_result.set_index('구별', inplace=True)
data_result.head()
```

## matplotlib의 CheetSheet

```python
import matplotlib.pyplot as plt

# data_result['CCTV대수']를 barh 형태로 plot 하자
plt.figure()
data_result['CCTV대수'].plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
# matplotlib으로 'CCTV대수'를 기주으로 가장 많은 구부터 출력
# sort_values를 data_result에 저장하고, 다음 줄에서 plot
data_result.sort_values(by='CCTV대수', ascending=False, inplace=True) # inplace를 적용해야 두줄에 나눠서 진행
data_result['CCTV대수'].plot(kind='bar', grid=True, figsize=(10, 10))
```

```python
data_result['CCTV대수'].sort_values(ascending=False).plot(kind='barh', grid=True, figsize=(10, 10))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구'], data_result['CCTV대수'], s=25)
plt.xlabel('인구')
plt.ylabel('CCTV대수')
plt.grid()
plt.show()
```

```python
data_result.head()
```

```python
# x, y 데이터를 받아들여서 1차 함수를 생성(직선의 방정식)
fp1 = np.polyfit(data_result['인구'], data_result['CCTV대수'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

# matplotlib으로 가로축은 '인구', 세로축은 'CCTV대수'인 scatter 그래프를 출력
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구'], data_result['CCTV대수'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
data_result['오차'] = np.abs(data_result['CCTV대수'] - f1(data_result['인구']))
```

```python
data_result['오차'] = np.abs(data_result['CCTV대수'] - f1(data_result['인구']))
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구'], data_result['CCTV대수'],
            c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(25):
    # text를 출력할 x좌표의 시작점, y좌표의 시작점, 찍을 내용
    plt.text(df_sort['인구'][n]*1.02, df_sort['CCTV대수'][n]*0.981,
             df_sort.index[n], fontsize=15)

temp = "Hello!!!"
plt.text(400000-25000, 2800, temp, fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
